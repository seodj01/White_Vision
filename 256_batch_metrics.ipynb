### Epoch Train/val Mix

```python
import matplotlib.pyplot as plt

# 첫 번째 데이터 (train)
epochs_train = new['epoch']
loss_train = new['epoch_loss']

# 두 번째 데이터 (validation)
epochs_val = val['epoch']
loss_val = val['epoch_loss']

# 그래프 크기 설정
plt.figure(figsize=(8, 5))

# Train Loss 선 그래프
plt.plot(epochs_train, loss_train, 
         color='blue', linestyle='-', linewidth=2, label='Train Loss')

# Validation Loss 선 그래프
plt.plot(epochs_val, loss_val, 
         color='red', linestyle='-', linewidth=2, label='Validation Loss')

# 축 레이블과 제목 설정
plt.xlabel('Epoch', fontsize=12, labelpad=10)
plt.ylabel('Loss', fontsize=12, labelpad=10)
plt.title('Train / Validation Loss (batch_size=256)', fontsize=14, pad=15)

# 격자 추가
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# 범례 추가
plt.legend(fontsize=10, loc='best')

# 그래프 출력
plt.show()
```
